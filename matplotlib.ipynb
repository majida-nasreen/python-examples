```python
import numpy as np
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
plt.subplot(2,1,1)
plt.plot(x,y,"r--")
plt.subplot(2,1,2)
plt.plot(y,x,"g*-");
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.4, 0.2, 0.4, 0.3])
axes1.plot(y,x,"#a83250")
axes2.plot(x,y,"r");
```

```python
fig, axes = plt.subplots()
axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title');
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

for ax in axes:
    ax.plot(x, y, 'g')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')

fig    
plt.tight_layout()
```

```python
x= np.linspace(1, 5,5)  

plt.figure(figsize=(5, 2.7), layout='constrained')
plt.plot(x, x, label='linear')  
plt.plot(x, x**2, label='quadratic')  
plt.plot(x, x**3, label='cubic')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend();
```
